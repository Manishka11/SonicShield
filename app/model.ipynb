```python
# SonicShield: Audio Threat Detection using AST and UrbanSound8K + ESC-50
# GPU-Optimized with Enhanced Progress Tracking

# 📌 Cell 1: Install Required Libraries
!pip install librosa torch torchvision torchaudio scikit-learn matplotlib seaborn tqdm --quiet

# 📌 Cell 2: Import Libraries
import os
import numpy as np
import pandas as pd
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.utils.class_weight import compute_class_weight

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torch.cuda.amp import autocast, GradScaler
from torchvision import transforms

from tqdm.auto import tqdm, trange
import time
import warnings
warnings.filterwarnings('ignore')

# Enable optimized CUDA operations
torch.backends.cudnn.benchmark = True
torch.backends.cudnn.deterministic = False

# 📌 Cell 3: Configuration & Threat Classes
THREAT_CLASSES = ['gun_shot', 'screaming', 'glass_breaking', 'siren', 'alarm']
SR = 22050
MAX_DURATION = 4.0
TARGET_LENGTH = int(SR * MAX_DURATION)

# GPU Memory Management
def print_gpu_memory():
    if torch.cuda.is_available():
        print(f"GPU Memory: {torch.cuda.memory_allocated()/1024**3:.2f}GB / {torch.cuda.memory_reserved()/1024**3:.2f}GB")

def clear_gpu_cache():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

# 📌 Cell 4: Optimized Preprocessing Class
class AudioPreprocessor:
    def __init__(self, sr=SR):
        self.sr = sr
        # Pre-compute mel filterbank for efficiency
        self.n_mels = 128
        self.hop_length = 512
        self.n_fft = 2048

    def load_audio(self, filepath):
        try:
            # Use faster loading with specified duration
            audio, _ = librosa.load(filepath, sr=self.sr, duration=MAX_DURATION, mono=True)
            if len(audio) < TARGET_LENGTH:
                audio = np.pad(audio, (0, TARGET_LENGTH - len(audio)), mode='constant')
            else:
                audio = audio[:TARGET_LENGTH]
            return audio.astype(np.float32)  # Use float32 for GPU efficiency
        except Exception as e:
            # Return silence if file can't be loaded
            return np.zeros(TARGET_LENGTH, dtype=np.float32)

    def extract_mel_spectrogram(self, audio):
        try:
            # Optimized mel-spectrogram extraction
            mel_spec = librosa.feature.melspectrogram(
                y=audio, 
                sr=self.sr, 
                n_mels=self.n_mels, 
                hop_length=self.hop_length,
                n_fft=self.n_fft,
                power=2.0
            )
            mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
            
            # Efficient normalization
            mean = mel_spec_db.mean()
            std = mel_spec_db.std()
            if std < 1e-6:
                std = 1.0
            
            normalized = (mel_spec_db - mean) / std
            return normalized.astype(np.float32)
        except Exception:
            # Return zero spectrogram on error
            return np.zeros((self.n_mels, 173), dtype=np.float32)

# 📌 Cell 5: Optimized Custom Dataset with Caching
class UrbanSoundDataset(Dataset):
    def __init__(self, file_paths, labels, preprocessor, cache_spectrograms=True):
        self.file_paths = file_paths
        self.labels = torch.LongTensor(labels)  # Pre-convert to tensor
        self.preprocessor = preprocessor
        self.cache_spectrograms = cache_spectrograms
        self.cache = {}
        
        # Pre-load spectrograms with progress bar if caching enabled
        if self.cache_spectrograms and len(self.file_paths) < 10000:  # Only cache if dataset is reasonable size
            self._preload_spectrograms()

    def _preload_spectrograms(self):
        print("Pre-loading spectrograms for faster training...")
        for idx in tqdm(range(len(self.file_paths)), desc="Caching spectrograms"):
            if self.file_paths[idx] != 'dummy':  # Skip dummy files
                try:
                    audio = self.preprocessor.load_audio(self.file_paths[idx])
                    mel_spec = self.preprocessor.extract_mel_spectrogram(audio)
                    self.cache[idx] = torch.FloatTensor(mel_spec).unsqueeze(0)
                except Exception:
                    self.cache[idx] = torch.zeros(1, 128, 173, dtype=torch.float32)

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        try:
            if self.cache_spectrograms and idx in self.cache:
                mel_spec = self.cache[idx]
            else:
                if self.file_paths[idx] == 'dummy':
                    # Generate random spectrogram for dummy data
                    mel_spec = torch.randn(1, 128, 173, dtype=torch.float32) * 0.1
                else:
                    audio = self.preprocessor.load_audio(self.file_paths[idx])
                    mel_spec_np = self.preprocessor.extract_mel_spectrogram(audio)
                    mel_spec = torch.FloatTensor(mel_spec_np).unsqueeze(0)
            
            return mel_spec, self.labels[idx]
        except Exception:
            # Return zero tensor and label 0 as fallback
            mel_spec = torch.zeros(1, 128, 173, dtype=torch.float32)
            return mel_spec, torch.tensor(0, dtype=torch.long)

# 📌 Cell 6: Load UrbanSound8K Dataset
def load_urbansound8k():
    # Check common Kaggle dataset paths
    possible_paths = [
        '/kaggle/input/urbansound8k',
        '/kaggle/input/urbansound8k-dataset',
        '../input/urbansound8k',
        '../input/urbansound8k-dataset'
    ]
    
    urb_path = None
    for path in tqdm(possible_paths, desc="Searching for UrbanSound8K"):
        if os.path.exists(path):
            urb_path = path
            break
    
    if urb_path is None:
        print("UrbanSound8K dataset not found. Please check dataset name in Kaggle.")
        return [], []
    
    print(f"✅ Found UrbanSound8K at: {urb_path}")
    
    # Check metadata file
    metadata_paths = [
        os.path.join(urb_path, 'metadata/UrbanSound8K.csv'),
        os.path.join(urb_path, 'UrbanSound8K.csv')
    ]
    
    metadata_path = None
    for path in metadata_paths:
        if os.path.exists(path):
            metadata_path = path
            break
    
    if metadata_path is None:
        print("❌ Metadata file not found")
        return [], []
    
    urb_meta = pd.read_csv(metadata_path)
    urb_files, urb_labels = [], []
    
    print("Loading UrbanSound8K files...")
    for _, row in tqdm(urb_meta.iterrows(), total=len(urb_meta), desc="Processing UrbanSound8K"):
        lbl = row['class'].lower().replace(' ', '_')
        # Try different audio folder structures
        audio_paths_to_try = [
            os.path.join(urb_path, f"audio/fold{row['fold']}/{row['slice_file_name']}"),
            os.path.join(urb_path, f"fold{row['fold']}/{row['slice_file_name']}"),
            os.path.join(urb_path, row['slice_file_name'])
        ]
        
        for p in audio_paths_to_try:
            if os.path.exists(p):
                urb_files.append(p)
                urb_labels.append(lbl)
                break
    
    print(f"✅ Loaded {len(urb_files)} UrbanSound8K files")
    return urb_files, urb_labels

# 📌 Cell 7: Load ESC-50 Dataset
def load_esc50():
    possible_paths = [
        '/kaggle/input/esc50',
        '/kaggle/input/esc-50',
        '../input/esc50',
        '../input/esc-50'
    ]
    
    esc_path = None
    for path in tqdm(possible_paths, desc="Searching for ESC-50"):
        if os.path.exists(path):
            esc_path = path
            break
    
    if esc_path is None:
        print("ESC-50 dataset not found.")
        return [], []
    
    print(f"✅ Found ESC-50 at: {esc_path}")
    
    # Check metadata file
    metadata_paths = [
        os.path.join(esc_path, 'meta/esc50.csv'),
        os.path.join(esc_path, 'esc50.csv'),
        os.path.join(esc_path, 'meta.csv')
    ]
    
    metadata_path = None
    for path in metadata_paths:
        if os.path.exists(path):
            metadata_path = path
            break
    
    if metadata_path is None:
        print("❌ ESC-50 metadata file not found")
        return [], []
    
    esc_meta = pd.read_csv(metadata_path)
    
    # Filter for relevant categories
    relevant_categories = ['screaming', 'glass_breaking', 'fire_alarm', 'dog_bark', 'gun_shot', 'siren']
    esc_meta = esc_meta[esc_meta['category'].isin(relevant_categories)]
    
    esc_files, esc_labels = [], []
    
    print("Loading ESC-50 files...")
    for _, row in tqdm(esc_meta.iterrows(), total=len(esc_meta), desc="Processing ESC-50"):
        audio_paths_to_try = [
            os.path.join(esc_path, 'audio', row['filename']),
            os.path.join(esc_path, row['filename'])
        ]
        
        for p in audio_paths_to_try:
            if os.path.exists(p):
                esc_files.append(p)
                esc_labels.append(row['category'].replace(' ', '_'))
                break
    
    print(f"✅ Loaded {len(esc_files)} ESC-50 files")
    return esc_files, esc_labels

# 📌 Cell 8: Load and Combine Datasets
print("🔍 Loading datasets...")
urb_files, urb_labels = load_urbansound8k()
esc_files, esc_labels = load_esc50()

file_paths = urb_files + esc_files
labels = urb_labels + esc_labels

if len(file_paths) == 0:
    print("⚠️ No audio files found! Creating dummy data for testing...")
    file_paths = ['dummy'] * 1000
    labels = (['gun_shot'] * 200 + ['screaming'] * 200 + ['glass_breaking'] * 200 + 
              ['siren'] * 200 + ['alarm'] * 200)

print(f"📊 Total files: {len(file_paths)}")
print(f"📊 Label distribution:\n{pd.Series(labels).value_counts()}")

encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)
class_names = encoder.classes_

print(f"🏷️ Classes: {list(class_names)}")

# Stratified split
X_train, X_test, y_train, y_test = train_test_split(
    file_paths, encoded_labels, 
    test_size=0.2, 
    stratify=encoded_labels if len(np.unique(encoded_labels)) > 1 else None,
    random_state=42
)

print(f"📈 Training samples: {len(X_train)}")
print(f"📉 Testing samples: {len(X_test)}")

# 📌 Cell 9: Create Optimized DataLoaders
print("🔧 Creating datasets and data loaders...")

preprocessor = AudioPreprocessor()

# Create datasets with caching
train_dataset = UrbanSoundDataset(X_train, y_train, preprocessor, cache_spectrograms=True)
test_dataset = UrbanSoundDataset(X_test, y_test, preprocessor, cache_spectrograms=True)

# Optimized DataLoader settings for GPU
batch_size = 32 if torch.cuda.is_available() else 16
num_workers = 4 if torch.cuda.is_available() else 2

train_loader = DataLoader(
    train_dataset, 
    batch_size=batch_size, 
    shuffle=True, 
    num_workers=num_workers,
    pin_memory=torch.cuda.is_available(),
    persistent_workers=True if num_workers > 0 else False
)

test_loader = DataLoader(
    test_dataset, 
    batch_size=batch_size, 
    shuffle=False, 
    num_workers=num_workers,
    pin_memory=torch.cuda.is_available(),
    persistent_workers=True if num_workers > 0 else False
)

print(f"✅ Created DataLoaders with batch_size={batch_size}, num_workers={num_workers}")

# 📌 Cell 10: Enhanced Model Architecture
class AudioTransformer(nn.Module):
    def __init__(self, num_classes, dropout_rate=0.3):
        super().__init__()
        
        # Convolutional layers with batch normalization
        self.conv1 = nn.Conv2d(1, 64, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(128)
        
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(256)
        
        # Adaptive pooling and dropout
        self.adaptive_pool = nn.AdaptiveAvgPool2d((4, 4))
        self.dropout1 = nn.Dropout(dropout_rate)
        
        # Fully connected layers
        self.fc1 = nn.Linear(256 * 4 * 4, 512)
        self.dropout2 = nn.Dropout(dropout_rate)
        self.fc2 = nn.Linear(512, 256)
        self.dropout3 = nn.Dropout(dropout_rate)
        self.fc3 = nn.Linear(256, num_classes)

    def forward(self, x):
        # Convolutional blocks
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.max_pool2d(x, 2)
        
        x = F.relu(self.bn2(self.conv2(x)))
        x = F.max_pool2d(x, 2)
        
        x = F.relu(self.bn3(self.conv3(x)))
        x = self.adaptive_pool(x)
        
        # Flatten and fully connected
        x = torch.flatten(x, 1)
        x = self.dropout1(x)
        
        x = F.relu(self.fc1(x))
        x = self.dropout2(x)
        
        x = F.relu(self.fc2(x))
        x = self.dropout3(x)
        
        return self.fc3(x)

# Setup device and model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"🖥️ Using device: {device}")

if torch.cuda.is_available():
    print(f"🚀 GPU: {torch.cuda.get_device_name()}")
    print_gpu_memory()

model = AudioTransformer(num_classes=len(class_names), dropout_rate=0.3).to(device)

# Model summary
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"📊 Model parameters: {total_params:,} total, {trainable_params:,} trainable")

# 📌 Cell 11: Optimized Training Loop
# Mixed precision training setup
scaler = GradScaler() if torch.cuda.is_available() else None

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)

# Learning rate scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3, factor=0.5, verbose=True)

# Training parameters
num_epochs = 10
best_val_acc = 0.0

print("🚀 Starting training...")
clear_gpu_cache()

training_history = {'train_loss': [], 'train_acc': [], 'val_acc': []}

for epoch in range(num_epochs):
    # Training phase
    model.train()
    train_loss = 0.0
    train_correct = 0
    train_total = 0
    
    # Create progress bar for training
    train_pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]')
    
    for batch_idx, (data, target) in enumerate(train_pbar):
        data, target = data.to(device, non_blocking=True), target.to(device, non_blocking=True)
        
        optimizer.zero_grad(set_to_none=True)  # More efficient than zero_grad()
        
        # Mixed precision forward pass
        if scaler is not None:
            with autocast():
                output = model(data)
                loss = criterion(output, target)
            
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
        
        # Statistics
        train_loss += loss.item()
        _, predicted = torch.max(output.data, 1)
        train_total += target.size(0)
        train_correct += (predicted == target).sum().item()
        
        # Update progress bar
        current_acc = 100. * train_correct / train_total
        train_pbar.set_postfix({
            'Loss': f'{loss.item():.4f}',
            'Acc': f'{current_acc:.2f}%',
            'LR': f'{optimizer.param_groups[0]["lr"]:.2e}'
        })
    
    # Calculate training metrics
    epoch_train_loss = train_loss / len(train_loader)
    epoch_train_acc = 100. * train_correct / train_total
    
    # Validation phase
    model.eval()
    val_correct = 0
    val_total = 0
    val_loss = 0.0
    
    val_pbar = tqdm(test_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Val]')
    
    with torch.no_grad():
        for data, target in val_pbar:
            data, target = data.to(device, non_blocking=True), target.to(device, non_blocking=True)
            
            if scaler is not None:
                with autocast():
                    output = model(data)
                    loss = criterion(output, target)
            else:
                output = model(data)
                loss = criterion(output, target)
            
            val_loss += loss.item()
            _, predicted = torch.max(output.data, 1)
            val_total += target.size(0)
            val_correct += (predicted == target).sum().item()
            
            # Update progress bar
            current_val_acc = 100. * val_correct / val_total
            val_pbar.set_postfix({'Acc': f'{current_val_acc:.2f}%'})
    
    epoch_val_acc = 100. * val_correct / val_total
    epoch_val_loss = val_loss / len(test_loader)
    
    # Update learning rate
    scheduler.step(epoch_val_loss)
    
    # Save best model
    if epoch_val_acc > best_val_acc:
        best_val_acc = epoch_val_acc
        torch.save(model.state_dict(), 'best_model.pth')
        print(f"💾 New best model saved! Validation accuracy: {best_val_acc:.2f}%")
    
    # Store history
    training_history['train_loss'].append(epoch_train_loss)
    training_history['train_acc'].append(epoch_train_acc)
    training_history['val_acc'].append(epoch_val_acc)
    
    # Print epoch summary
    print(f"📊 Epoch {epoch+1}/{num_epochs} Summary:")
    print(f"   Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.2f}%")
    print(f"   Val Acc: {epoch_val_acc:.2f}%, Best Val Acc: {best_val_acc:.2f}%")
    
    if torch.cuda.is_available():
        print_gpu_memory()
    
    clear_gpu_cache()

print(f"🎉 Training completed! Best validation accuracy: {best_val_acc:.2f}%")

# 📌 Cell 12: Enhanced Evaluation with Threat Detection
print("🔍 Loading best model and evaluating...")

# Load best model
model.load_state_dict(torch.load('best_model.pth'))
model.eval()

# Threat detection results
threat_detections = []
all_predictions = []
all_targets = []

print("\n🚨 Real-time Threat Detection:")
eval_pbar = tqdm(test_loader, desc="Evaluating & Detecting Threats")

with torch.no_grad():
    for batch_idx, (data, target) in enumerate(eval_pbar):
        data, target = data.to(device, non_blocking=True), target.to(device, non_blocking=True)
        
        if scaler is not None:
            with autocast():
                output = model(data)
        else:
            output = model(data)
        
        probabilities = F.softmax(output, dim=1)
        predictions = output.argmax(1)
        
        # Process each sample in batch
        for i in range(len(predictions)):
            pred_class = encoder.inverse_transform([predictions[i].item()])[0]
            confidence = probabilities[i][predictions[i]].item()
            
            # Threat detection with different thresholds
            if pred_class in THREAT_CLASSES:
                if confidence > 0.8:
                    threat_level = "🔴 HIGH"
                elif confidence > 0.6:
                    threat_level = "🟡 MEDIUM"
                else:
                    threat_level = "🟢 LOW"
                
                if confidence > 0.6:  # Only report medium and high confidence
                    detection_info = {
                        'class': pred_class,
                        'confidence': confidence,
                        'level': threat_level
                    }
                    threat_detections.append(detection_info)
                    
                    tqdm.write(f"{threat_level} THREAT: {pred_class.upper()} "
                             f"(Confidence: {confidence:.3f})")
        
        all_predictions.extend(predictions.cpu().numpy())
        all_targets.extend(target.cpu().numpy())
        
        # Update progress bar with threat count
        eval_pbar.set_postfix({'Threats Detected': len(threat_detections)})

# 📌 Cell 13: Comprehensive Results Report
print("\n" + "="*60)
print("📊 COMPREHENSIVE EVALUATION REPORT")
print("="*60)

if len(all_predictions) > 0:
    # Overall accuracy
    final_accuracy = accuracy_score(all_targets, all_predictions)
    print(f"\n🎯 Overall Test Accuracy: {final_accuracy:.3f} ({final_accuracy*100:.1f}%)")
    
    # Detailed classification report
    print("\n📋 Detailed Classification Report:")
    report = classification_report(all_targets, all_predictions, 
                                 target_names=class_names, 
                                 zero_division=0,
                                 digits=3)
    print(report)
    
    # Threat detection summary
    print(f"\n🚨 Threat Detection Summary:")
    print(f"   Total threats detected: {len(threat_detections)}")
    
    if threat_detections:
        threat_df = pd.DataFrame(threat_detections)
        threat_counts = threat_df['class'].value_counts()
        print("   Threats by class:")
        for threat_class, count in threat_counts.items():
            avg_conf = threat_df[threat_df['class'] == threat_class]['confidence'].mean()
            print(f"     {threat_class}: {count} detections (avg confidence: {avg_conf:.3f})")
    
    # Confusion Matrix
    cm = confusion_matrix(all_targets, all_predictions)
    plt.figure(figsize=(14, 10))
    sns.heatmap(cm, annot=True, fmt='d', 
                xticklabels=class_names, yticklabels=class_names, 
                cmap='Blues', cbar_kws={'label': 'Count'})
    plt.title("Confusion Matrix - Audio Threat Detection System", fontsize=16, fontweight='bold')
    plt.xlabel("Predicted Class", fontsize=12)
    plt.ylabel("True Class", fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
    
    # Training history plot
    if training_history['train_loss']:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
        
        # Loss plot
        ax1.plot(training_history['train_loss'], 'b-', label='Training Loss', linewidth=2)
        ax1.set_title('Training Loss Over Time', fontweight='bold')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Accuracy plot
        ax2.plot(training_history['train_acc'], 'b-', label='Training Accuracy', linewidth=2)
        ax2.plot(training_history['val_acc'], 'r-', label='Validation Accuracy', linewidth=2)
        ax2.set_title('Model Accuracy Over Time', fontweight='bold')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy (%)')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()

else:
    print("❌ No predictions made - check dataset loading and model training.")

# Final GPU memory cleanup
clear_gpu_cache()

print("\n✅ SonicShield Analysis Complete!")
print(f"🏆 Best Model Accuracy: {best_val_acc:.2f}%")
print(f"🔍 Total Threat Detections: {len(threat_detections)}")
print("="*60)
```
